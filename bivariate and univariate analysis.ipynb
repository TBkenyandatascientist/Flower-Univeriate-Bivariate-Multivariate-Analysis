```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import seaborn as sns
```

```python
# loading data  file
data = pd.read_csv('iris.data', sep=',', names=["sepal_length", "sepal_width" ,"petal_length", "petal_width", "class"])

data.head()
```

### univariate analysis

```python
data.head()
```

```python
data["class"].unique()
```

```python
#creating a variable that represents the setosa class
setosa_class = data.loc[data["class"]== "Iris-setosa"]
setosa_class
```

```python
#creating second group with second class
versicolor_class = data.loc[data["class"]=="Iris-versicolor"]
versicolor_class
```

```python
#creating third group with third class
virginica_class = data.loc[data["class"]=="Iris-virginica"]
virginica_class
```

```python
plt.plot(setosa_class["sepal_length"], np.zeros_like(setosa_class["sepal_length"]), 'o')
plt.plot(virginica_class["sepal_length"], np.zeros_like(virginica_class["sepal_length"]), 'o')
plt.plot(versicolor_class["sepal_length"], np.zeros_like(versicolor_class["sepal_length"]), 'o')
plt.xlabel("sepal_length")
plt.show()
```

### bivariate analysis

```python
sns.FacetGrid(data, hue="class", size=5).map(plt.scatter, "sepal_length", "sepal_width").add_legend();
plt.show()
```

```python
#reflection
Iris setosa is easily identifiable while iris-versicolor and iris-virginica tend to overlap.
no correlation was detected between sepal_length and sepal_width
```

```python
sns.FacetGrid(data, hue="class", size=5).map(plt.scatter, "petal_length", "petal_width").add_legend();
plt.show()
```

```python
#reflection
there was a positive correlation between the petal_length and petal_width and therefore these two 
features are the most useful features to highlight the different types of flowers.
Iris setosa is easily identifiable while iris-versicolor and iris-virginica tend to overlap.
```
